```python
years = sorted(list(summer_athlete['Year'].unique()))
sports = []
events = []
teams = []
athlete_number = []
females = []
males = []
for year in years:
    df = summer_athlete.loc[summer_athlete['Year'] == year]
    
    sport = list(df['Sport'].unique())
    sports.append(len(sport))
    
    team = list(df['NOC'].unique())
    teams.append(len(team))
    
    event = list(df['Event'].unique())
    events.append(len(event))
    
    number = list(df['ID'].unique())
    athlete_number.append(len(number))
    
    sex = df.drop_duplicates(subset=['ID'], keep='first')
    sex_f = sex.loc[sex['Sex'] == 'F']
    female = list(sex_f['ID'].unique())
    females.append(len(female))
    
    sex_m = sex.loc[sex['Sex'] == 'M']
    male = list(sex_m['ID'].unique())
    males.append(len(male))
    
summer = pd.DataFrame({'Sports':sports, 'Teams':teams, 'Event':events, 
                       'Number of Athlete':athlete_number, 'Females':females, 'Males': males})
summer.index = years
```

```python
df2014 = winter_athlete.loc[winter_athlete['Year'] == 2014]
team2014 = list(df2014['NOC'].unique())
athlete2014 = []
country2014 = []
for noc in team2014:
    df = df2014.loc[df2014['NOC'] == noc]
    number = list(df['ID'].unique())
    athlete2014.append(len(number))
    name2014 = noc_to_country['region'].loc[noc_to_country['NOC'] == noc]
    country2014.append(name2014.values[0])
num2014 = pd.DataFrame(list(zip(country2014, athlete2014)), columns =['Country', 'Number of Athlete']) 
num2014 = num2014.set_index('Country')
num2014 = num2014.groupby('Country').sum()
```

```python
def count_noc(noc,season):
    sports = []
    events = []
    athlete_number = []
    females = []
    males = []
    b = pd.DataFrame(index=['Gold','Silver','Bronze'])
    df1 = season.loc[season['NOC'] == noc]
    years = sorted(list(df1['Year'].unique()))
    for year in years:
        df = df1.loc[df1['Year'] == year]

        sport = list(df['Sport'].unique())
        sports.append(len(sport))

        event = list(df['Event'].unique())
        events.append(len(event))

        number = list(df['ID'].unique())
        athlete_number.append(len(number))

        sex = df.drop_duplicates(subset=['ID'], keep='first')
        sex_f = sex.loc[sex['Sex'] == 'F']
        female = list(sex_f['ID'].unique())
        females.append(len(female))

        sex_m = sex.loc[sex['Sex'] == 'M']
        male = list(sex_m['ID'].unique())
        males.append(len(male))

        aa = df.drop_duplicates(subset=['Event'], keep='first')
        a = pd.DataFrame(aa['Medal'].value_counts())
        b = pd.concat([b,a],axis=1)
        
    b.columns  = years
    b = b.fillna(int(0))
    b.loc['Total Medal'] = b.apply(lambda x: x.sum())
    NOC = pd.DataFrame({'Sports':sports, 'Event':events, 'Number of Athlete':athlete_number,
                        'Females':females, 'Males': males,
                       'Gold':b.loc['Gold',:], 'Silver':b.loc['Silver',:], 'Bronze':b.loc['Bronze',:],
                        'Total Medal':b.loc['Total Medal',:]})
    NOC.index = years
    return NOC
```
